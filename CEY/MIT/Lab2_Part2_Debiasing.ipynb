공정하고 편견이 없는 AI 시스템을 구축하는 것은 장기적인 수용에 매우 중요하다. 얼굴 감지의 과제를 고려해보라: 주어진 이미지는 얼굴의 이미지인가? 간단해 보이지만 매우 중요한 이 작업은 선택된 인구통계학 사이에서 상당한 양의 알고리즘 편향을 받는다.

이 실험실에서, 우리는 알고리즘 편향을 해결하기 위해 최근에 발표된 한 가지 접근법을 조사할 것이다. 얼굴 이미지 데이터 세트의 기본이 되는 잠재 변수를 학습하고 이를 사용하여 훈련 데이터를 적응적으로 재샘플링하여 저하된 모델을 훈련하기 위해 존재할 수 있는 편견을 완화하는 안면 감지 모델을 구축할 것이다.
